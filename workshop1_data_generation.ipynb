```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 1. Setup & Seed
np.random.seed(42)
random.seed(42)

# กำหนดตัวแปรเวลาปัจจุบัน (ปี 2026 ตามที่ต้องการ)
current_date = datetime(2026, 4, 3)
three_months_later = current_date + timedelta(days=90)

# รายชื่อวัตถุดิบ 20 รายการ
rm_names = [f"RM{i:02d}" for i in range(1, 21)]
products = [f"Product_{chr(65+i)}" for i in range(5)]

# ฟังก์ชันสุ่มวันที่ระหว่างช่วง และแปลงเป็น Format YYYYMMDD
def random_date_str(start, end):
    delta = end - start
    random_days = random.randint(0, delta.days)
    return (start + timedelta(days=random_days)).strftime('%Y%m%d')

# --- 2. สร้างข้อมูล 01-inventory.csv ---
stock_quantities = []
for _ in range(20):
    if random.random() < 0.4: # 40% มีของน้อย
        stock_quantities.append(np.random.randint(5, 40))
    else: # 60% มีปกติ
        stock_quantities.append(np.random.randint(100, 500))

df_inventory = pd.DataFrame({
    'idx': range(1, 21),
    'rm_name': rm_names,
    'stock_qty': stock_quantities,
    'unit': ['ton'] * 20,
    'stock_price': np.random.randint(400, 10001, size=20),
    'eco_score': np.round(np.random.uniform(0, 10, size=20), 2)
})
df_inventory.to_csv('01-inventory.csv', index=False)

# --- 3. สร้างข้อมูล 02-recipe.csv ---
recipe_rows = []
for prod in products:
    num_recipes = random.randint(1, 3)
    prod_suffix = prod.split('_')[-1] # ดึง A, B, C...

    for r_idx in range(1, num_recipes + 1):
        recipe_id = f"RECIPE-{prod_suffix}-{r_idx}" # Format: RECIPE-A-1
        num_rms = random.randint(3, 5)
        selected_rms = random.sample(rm_names, num_rms)

        # ใช้ Dirichlet สุ่มให้ผลรวมเป็น 1.0 (100%)
        props = np.random.dirichlet(np.ones(num_rms))

        for rm, p in zip(selected_rms, props):
            recipe_rows.append({
                'recipe_number': recipe_id,
                'product_name': prod,
                'rm_name': rm,
                'usage_percent': round(p, 4)
            })

df_recipe = pd.DataFrame(recipe_rows)
df_recipe.insert(0, 'idx', range(1, len(df_recipe) + 1))
df_recipe.to_csv('02-recipe.csv', index=False)

# --- 4. สร้างข้อมูล 03-orders.csv ---
order_list = []
for i in range(1, 21):
    mfg_dt = random_date_str(current_date, three_months_later)
    # due_date = mfg_date + 5 วัน
    due_dt = (datetime.strptime(mfg_dt, '%Y%m%d') + timedelta(days=5)).strftime('%Y%m%d')

    order_list.append({
        'idx': i,
        'mo_number': f"MO{i:03d}",
        'product_name': random.choice(products),
        'mfg_date': mfg_dt,
        'due_date': due_dt,
        'order_qty': np.random.randint(10, 100)
    })
df_order = pd.DataFrame(order_list)
df_order.to_csv('03-orders.csv', index=False)

# --- 5. สร้างข้อมูล 04-git.csv ---
git_list = []
for i in range(1, 21):
    git_list.append({
        'idx': i,
        'git_number': f"GIT{i:03d}",
        'rm_name': random.choice(rm_names),
        'git_qty': np.random.randint(50, 200),
        'delivery_date': random_date_str(current_date, three_months_later)
    })
df_git = pd.DataFrame(git_list)
df_git.to_csv('04-git.csv', index=False)

print("✅ All 4 files created successfully with YYYYMMDD format and correct naming conventions.")
```
